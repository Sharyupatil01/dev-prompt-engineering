```python
import requests 
```

```python
prompt="write a one sentence for unicorn sleeping at night"

def generate_response(temp=None):
    url = "http://localhost:11434/api/generate"

    payload = {
        "model": "phi",
        "prompt": prompt,
        "stream": False
    }

    #Add temperature in payload if not provided 
    if temp is not None:
        payload["options"]={
            "temperature":temp
        }
    
    response=requests.post(url,json=payload)
    return response.json()["response"]
```

```python
#first response 
print(f"prompt:{prompt}")
print(f"response:{generate_response()}")

#second response with temperature 

print("\nWith Controlled Response parameters: ")
print(f"response with controlled temperatures :{generate_response(temp=0.7)}")
```
