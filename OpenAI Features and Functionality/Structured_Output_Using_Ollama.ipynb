```python
import requests 
import json #to generate the AI ouput -> python dict 
```

```python


# Step 1: Ask Ollama to generate UI JSON
prompt = """
You are a UI generator AI.

Return ONLY valid JSON in this format:
{
  "type": "form",
  "label": "User Form",
  "children": [
    {
      "type": "field",
      "label": "Name"
    },
    {
      "type": "button",
      "label": "Submit"
    }
  ]
}

User request: Create a simple user form
"""

response = requests.post(
    "http://localhost:11434/api/generate",
    json={
        "model": "phi",
        "prompt": prompt,
        "stream": False,
        "options": {"temperature": 0.3}
    }
)

# Step 2: Convert JSON string → Python dict
ui = json.loads(response.json()["response"])

# Step 3: Use the dict (render simple HTML)
def render(ui):
    if ui["type"] == "form":
        print(f"\n<form>")
        for child in ui["children"]:
            render(child)
        print("</form>")

    elif ui["type"] == "field":
        print(f"<label>{ui['label']}</label>")
        print(f"<input placeholder='{ui['label']}' />")

    elif ui["type"] == "button":
        print(f"<button>{ui['label']}</button>")

# Step 4: Display UI
render(ui)
```
